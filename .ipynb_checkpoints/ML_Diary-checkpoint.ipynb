```python
from scipy import ndimage
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("1.jpg")
img = img[500: 850, 100:375]
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap='gray')
```

```python
def view(img):
    plt.figure(figsize=(10, 10))
    plt.imshow(img, cmap='gray')
```

```python
def find_hline(image, length, col=0):
    
    l = []
    for i in range(image.shape[1]):
        if image[col, i:i+length].sum() == 0:
            l.append(i)
    
    if l == []:
        return None
    else:
        line = [l[0], col, l[-1], col]
    
    return line

def hlines(image, length):
    
    lines = []
    for i in range(image.shape[0]):
        new_line = find_hline(image, length, i)
        if new_line == None:
            continue
        else: 
            lines.append(new_line)
            
    return lines
```

```python
thresh1 = cv2.adaptiveThreshold(img_gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 3)
view(thresh1)
```

```python
thresh1 = thresh1[0:350, 0:252]
hl = hlines(thresh1, 20)
```

```python
img_test = np.full(thresh1.shape, 255, dtype='uint8')

for line in hl:
    cv2.line(img_test, (line[0], line[1]), (line[2], line[3]), (0,0,0), 2)
    
view(img_test)
```

```python
def strong_horisontal(image, h, l, long=10): # h - hight of window, l - length of window
    
#     находим сгустки
    lines = []
    for i in range(image.shape[0]):
        line = []
        for j in range(image.shape[1]):
            if image[i:i+h, j:j+l].sum() == 0:
                line.append(j)
            else:
                continue
            lines.append([i, line[0], i, line[-1]])
    
#     оставляем длинные
    long_lines = []
    for l in lines:
        if l[3] - l[1] > long:
            long_lines.append(l)
            
#     сохраняем координату горизонтали

    rows = []
    for l in long_lines:
        if l[0] in rows:
            continue
        else:
            rows.append(l[0])
    
    strong_lines = []
    for i in range(len(rows)):
        if i == 0:
            strong = [rows[i]]
        elif rows[i] - rows[i-1] < 10:
            strong.append(rows[i])
        else:
            strong_lines.append(strong)
            strong = [rows[i]]
    strong_lines.append(strong)
            
    maxi = image.shape[1]
    
    final_lines = []
    for i in strong_lines:
        i = round(np.array(i).mean())
        final_lines.append([i, 0, i, maxi])
            
    return final_lines
    
    
```

```python
strhLines = strong_horisontal(img_test, 2, 10, 5)
strhLines = np.array(strhLines, dtype = 'int')
strhLines
```

```python
img_final = np.full(thresh1.shape, 255, dtype='uint8')

for line in strhLines:
    cv2.line(img_final, (line[1], line[0]), (line[3], line[2]), (0,0,0), 1)
    
view(img_final)
```

```python
view(thresh1.T)

vl = hlines(thresh1.T, 20)
img_test_v = np.full(thresh1.T.shape, 255, dtype='uint8')

for line in vl:
    cv2.line(img_test_v, (line[0], line[1]), (line[2], line[3]), (0,0,0), 2)
    
view(img_test_v)

str_vert = strong_horisontal(img_test_v, 2, 10, 5)
str_vert = np.array(str_vert, dtype = 'int')

img_vert = np.full(thresh1.T.shape, 255, dtype='uint8')

for line in str_vert:
    cv2.line(img_vert, (line[1], line[0]), (line[3], line[2]), (0,0,0), 1)
    
view(img_vert)
```

```python
for line in str_vert:
    cv2.line(img_final, (line[0], line[1]), (line[2], line[3]), (0,0,0), 1)
    
view(img_final)
```
